## Repeat 50 times

```python
def mse_list():
        
    #initialise the model
    model = regression_model()
    
    for i in range(50):
        #step 1
        X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=1234)
        #step 2
        model.fit(X_train, y_train, epochs=50)
        #step 3
        y_pred = model.predict(X_test)
        MSE = mean_squared_error(y_test, y_pred)
        MSE_list.append(MSE)
```
